```python
# importing the required module 
import matplotlib.pyplot as plt 
```

```python
# x axis values 
x = [1,2,3] 
# corresponding y axis values 
y = [2,4,1] 
```

```python
print(x, y)
```

```python
# plotting the points  
plt.plot(x, y)
# naming the x axis 
plt.xlabel('age') 
# naming the y axis 
plt.ylabel('salary') 
# giving a title to my graph 
plt.title('Age vs Salary') 
# function to show the plot 
plt.show() 
```

```python
import pandas as pd
import numpy as np
from pydataset import data
```

```python
mtcars = data('mtcars')
mtcars.shape
```

```python
mtcars.columns
```

```python
mtcars[['wt','mpg']].sort_values(by='wt').head()
```

```python
mtcars['am'] = mtcars['am'].astype('category')
mtcars.dtypes
```

```python
mtcars.am.value_counts()
```

```python
mtcars.plot.scatter(x='wt', y='mpg', s='hp', c='am', colormap='viridis')
```

```python
#mpl
matplotlib.pyplot.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, *, edgecolors=None, plotnonfinite=False, data=None, **kwargs)[source]
plt.scatter(x='wt', y='mpg', data=mtcars)
```

```python
plt.scatter(x='wt', y='mpg', data=mtcars, s='hp', c='am', marker='2')
```

```python
#plt.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)[source]
gears = mtcars.gear.value_counts().reset_index()
gears
```

```python
plt.bar(x='gear',height = 'count', data=gears)
```

```python
carType = mtcars.groupby('am').agg({'gear':'count'}).reset_index().rename(columns={'gear':'count'})
carType
```

```python
plt.bar(x='am',height = 'count', data=carType)
```
