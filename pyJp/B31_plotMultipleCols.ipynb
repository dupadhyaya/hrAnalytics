```python
# figure and axes

fig, ax = plt.subplots(1, figsize=(6,8))  #set figure size also
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Title')
plt.xlim(0,5)
plt.ylim(0,20)
plt.plot([1,1,2,3,3,4], [1,3,4,9,15,19], linestyle='--', color='r')
plt.scatter([1,2,3,4], [1,4,9,19], color='b')
plt.grid(True)
plt.text(1, 3, 'Hello at (1, 3)')
plt.annotate('Hello at 3, 15', xy=(0.2,.5), xytext=(3,15))
arrowprops = dict(facecolor='black', shrink=.5)
plt.savefig('figure.png')  #before show
plt.show();
```
