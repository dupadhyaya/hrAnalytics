# sklearn

```python
from sklearn import datasets
```

```python
data = datasets.load_wine(return_X_y = False, as_frame=True)
data = data['frame']
data.head()
```

```python
data, target = datasets.load_iris(return_X_y = True, as_frame= True)
data['target'] = target
data.head()
```

```python
data = datasets.fetch_california_housing(return_X_y=False, as_frame=True)
data = data['frame']
data.head()
```

```python
data = datasets.fetch_openml('diabetes', parser='auto', version=1, as_frame=True, return_X_y=False)
data = data['frame']
data.head()
```

```python
data = datasets.fetch_openml(data_id=42437, parser='auto',  as_frame=True, return_X_y=False)
data = data['frame']
data.head()
```
