```python
#libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pydataset import data
import plotnine
import os
```

```python
#add libraries
from matplotlib.backends.backend_pdf import PdfPages
```

```python
df = data('mtcars')
```

```python
df.head()
```

### Convert the pandas dataframe to a table using matplotlib

```python
fig, ax =plt.subplots(figsize=(12,4))
ax.axis('tight')
ax.axis('off')
the_table = ax.table(cellText=df.head(20).values,colLabels=df.columns,loc='center')
```

### Save the Pdf File that have the pandas dataframe

```python
pp = PdfPages("../doc/mtcars.pdf") #pl have folder doc inside ur repository
pp.savefig(fig, bbox_inches='tight')
pp.close();
```
