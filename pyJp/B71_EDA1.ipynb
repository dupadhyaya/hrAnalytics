```python
#compare more features
g = sns.pairplot(data=df3, vars=['Year_Introduced', 'Speed_mph', 'Height_ft', 'Inversions', 'Gforce'])
#g.fig.suptitle('Pair Plot')
plt.suptitle('Main Title : Pair Plot', y=1.01)
plt.show();
```

# Step - 5 : Questions about Data
-  1 What are the locations with the fastest roller coasters ?

```python
df3.head(2)
```

```python
df3['Location'].value_counts()
```

```python
ax = df3.query('Location != "Other"') \
.groupby('Location')['Speed_mph'] \
.agg(['mean','count']) \
.query('count >= 10') \
.sort_values('mean') ['mean']\
.plot(kind='barh', figsize=(12,5), title='Average Coast Speed by Location')

ax.set_xlabel('Average Coaster Speed')
plt.show();
```
