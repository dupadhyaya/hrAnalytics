```python
#altair
import altair as alt

# load a simple dataset as a pandas DataFrame
from vega_datasets import data
cars = data.cars()
print(cars.head())
```

```python
alt.Chart(cars).mark_point().encode( x='Horsepower',    y='Miles_per_Gallon',     color='Origin' )
```

```python
import altair as alt
mtcars[['gear','cyl','carb', 'am','vs']] = mtcars[['gear','cyl','carb', 'am','vs']].astype('category')
alt.Chart(mtcars).mark_point().encode(  x='wt', y='mpg', color='gear',)
```

```python
#plotNine
from plotnine import *
```

```python
(ggplot(mtcars, aes("wt", "mpg")) + geom_point())
```

```python
(ggplot(mtcars, aes("wt", "mpg", color="factor(gear)"))  + geom_point())
```
