# C23 Graphs Matplotlib & Seaborn
-  matplotlib.pyplot is a collection of functions that make matplotlib work like MATLAB. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.
-  https://matplotlib.org/stable/api/pyplot_summary.html

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pydataset import data
```

```python
print(sns.__version__)
```

```python
mtcars = data('mtcars')
```

```python
mtcars
```

## Scatter Plot

```python
plt.scatter(mtcars.wt, mtcars.mpg, color='r', s=10)
plt.show();
```

```python
sns.scatterplot(x='wt', y='mpg', data=mtcars)
plt.show();
```

##  Line Plot

```python
plt.plot(mtcars.wt, mtcars.mpg)
plt.show()
```

```python
mtcarsSort = mtcars.sort_values(by='wt')
mtcarsSort.head()
```

```python
plt.plot(mtcarsSort.wt, mtcarsSort.mpg, color='r')
plt.show();
```

```python
plt.plot(mtcarsSort[['wt']], mtcarsSort[['mpg']], color='r', label='MPG', linestyle='dashed')
plt.plot(mtcarsSort[['wt']], mtcarsSort[['hp']], color='g', label='HP', linewidth=.4, marker='x')
plt.legend()
plt.ylabel('MPG and HP')
plt.xlabel('Wt')
plt.title('Wt vs MPG- HP')
plt.grid()
plt.show();
```

## Pie Plot

```python
mtcars['gear'].unique()
```

```python
plt.pie(x=mtcars.gear.value_counts(), labels=mtcars.gear.unique())
plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(mtcars.gear.value_counts(), labels=mtcars.gear.unique(), autopct='%1.1f%%', radius=1.2)
plt.show()
```

## Correlation Plot

```python
mtcars.corr().round(2)
```

```python
sns.heatmap(mtcars[['wt','mpg','hp']].corr().round(2), annot=True, vmin=-1, vmax=1 )
plt.show()
```

## Bar Plot

```python
gearCnt = mtcars.groupby('gear').size().reset_index().rename(columns={0:'count'})
gearCnt
```

```python
plt.bar(x='gear', height='count', data=gearCnt)
plt.show()
```

```python
plt.barh(y='gear', width='count', data=gearCnt, color='green')
plt.show()
```

```python
sns.barplot(x='gear', y='count', data=gearCnt)
plt.show()
```

## Area Plot

```python
months = range(1, 13)
sale = [14, 13, 10, 15, 17, 15, 15, 13, 12, 10, 14, 11]
plt.fill_between(months, sale, color='darkorange')
plt.show();
```

```python
months = range(1, 13)
sale = [14, 13, 10, 15, 17, 15, 15, 13, 12, 10, 14, 11]
ad = [13, 17, 12, 11, 11, 10, 15, 14, 13, 12, 11, 15]
plt.stackplot(months, sale, ad, labels=['Sales','Ads'])
plt.legend()
plt.show();
```

## Heat Map

```python
corrData = mtcars[['wt','mpg', 'hp']].corr().round(2)
corrData
```

```python
plt.imshow(corrData, cmap='autumn')
plt.colorbar()
plt.show()
```

```python
sns.heatmap(data=corrData, annot=True)
plt.show();
```

## Histogram Plot

```python
plt.hist(x=mtcars.mpg)
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)

# We can set the number of bins with the *bins* keyword argument.
axs[0].hist(mtcars.wt, bins=5, histtype='bar', edgecolor='black')
axs[1].hist(mtcars.mpg, bins=10, histtype='step')
```

```python
sns.histplot(data=mtcars['mpg'], kde=True, color='lightgreen', edgecolor='r')
```

```python
counts, bins = np.histogram(mtcars.mpg)
plt.stairs(counts, bins)
plt.show()
```

## Box Plot

```python
plt.boxplot(x=mtcars.mpg, notch=True, )
plt.show()
```

## Violin Plot

```python
plt.violinplot(dataset=mtcars[['mpg','wt']], widths=.3, showmeans=True )
plt.show()
```
